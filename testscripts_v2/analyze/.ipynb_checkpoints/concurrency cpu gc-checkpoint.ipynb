```python
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import pandas as pd
import numpy as np
import os
import sys
```

```python
df1 = pd.read_csv('results_20190920_083532.log', sep=',')
```

```python
df1 = df1.drop(columns=['jvm_short','duration'])
```

```python
#Cleaning data and setting the correct datatypes
df1=df1.loc[df1['failed_req']!='FAILED']
df1['failed_req']=pd.to_numeric(df1['failed_req'])
df1['compl_req']=pd.to_numeric(df1['compl_req'])
df1['req_per_sec']=pd.to_numeric(df1['req_per_sec'])
df1['time_per_req_avg']=pd.to_numeric(df1['time_per_req_avg'])
df1['cpus']=pd.to_numeric(df1['cpus'])
df1['concurrency']=pd.to_numeric(df1['concurrency'])
```

```python
# which framework, JVM, GC algorithm can deal best with more CPUs and with higher concurrency. throughput
df1=df1.loc[(df1['jvm_vendor']=='OpenJDK') & (df1['jvm_major_version']==8) & (df1['gc_short']=='G1GC') & (df1['memflag']=='-Xmx128m -Xms128m')]
#print(df1.head(5))
frameworks=df1.framework.unique()
gcs=df1.gc_short.unique()
jvms=df1.jvm_vendor.unique()

df = pd.DataFrame(columns=['framework','tru_cpu_impr','tru_conc_impr','tru_impr','avg_cpu_impr','avg_conc_impr','avg_impr' ])
for framework in frameworks:
    framework_tru_1_1=df1.loc[(df1['framework']==framework) & (df1['concurrency']==1) & (df1['cpus']==1)].iloc[0]['compl_req']
    framework_tru_1_4=df1.loc[(df1['framework']==framework) & (df1['concurrency']==1) & (df1['cpus']==4)].iloc[0]['compl_req']
    framework_tru_1_8=df1.loc[(df1['framework']==framework) & (df1['concurrency']==1) & (df1['cpus']==8)].iloc[0]['compl_req']
    framework_tru_4_8=df1.loc[(df1['framework']==framework) & (df1['concurrency']==4) & (df1['cpus']==8)].iloc[0]['compl_req']
    framework_tru_4_1=df1.loc[(df1['framework']==framework) & (df1['concurrency']==4) & (df1['cpus']==1)].iloc[0]['compl_req']
    framework_tru_4_4=df1.loc[(df1['framework']==framework) & (df1['concurrency']==4) & (df1['cpus']==4)].iloc[0]['compl_req']
    framework_avg_1_1=df1.loc[(df1['framework']==framework) & (df1['concurrency']==1) & (df1['cpus']==1)].iloc[0]['time_per_req_avg']
    framework_avg_1_4=df1.loc[(df1['framework']==framework) & (df1['concurrency']==1) & (df1['cpus']==4)].iloc[0]['time_per_req_avg']
    framework_avg_1_8=df1.loc[(df1['framework']==framework) & (df1['concurrency']==1) & (df1['cpus']==8)].iloc[0]['time_per_req_avg']
    framework_avg_4_8=df1.loc[(df1['framework']==framework) & (df1['concurrency']==4) & (df1['cpus']==8)].iloc[0]['time_per_req_avg']
    framework_avg_4_1=df1.loc[(df1['framework']==framework) & (df1['concurrency']==4) & (df1['cpus']==1)].iloc[0]['time_per_req_avg']
    framework_avg_4_4=df1.loc[(df1['framework']==framework) & (df1['concurrency']==4) & (df1['cpus']==4)].iloc[0]['time_per_req_avg']
    
    #impr = framework_val_1_1/framework_val_4_8
    tru_cpu_impr = 100*framework_tru_1_8/framework_tru_1_1
    tru_conc_impr = 100*framework_tru_4_1/framework_tru_1_1
    avg_cpu_impr = 100*framework_avg_1_8/framework_avg_1_1
    avg_conc_impr = 100*framework_avg_4_1/framework_avg_1_1
    
    tru_impr = 100*framework_tru_4_4/framework_tru_1_1
    avg_impr = 100*framework_avg_4_4/framework_avg_1_1

    
    #conc_impr = 100*framework_val_1_1/framework_val_4_4

    #print (str(framework_val_1_1)+"\t"+str(framework_val_4_8))
    df = df.append({'framework' : framework , 'tru_cpu_impr' : tru_cpu_impr, 'tru_conc_impr' : tru_conc_impr, 'tru_impr' : tru_impr,'avg_cpu_impr' : avg_cpu_impr, 'avg_conc_impr' : avg_conc_impr, 'avg_impr' : avg_impr} , ignore_index=True)

df=df.sort_values(['tru_conc_impr'], ascending=[False])
figure(num=None, figsize=(16, 6))

y_pos = np.arange(len(frameworks))
plt.bar(y_pos, df['tru_conc_impr'].values, align='center', alpha=0.5)
plt.xticks(y_pos, df['framework'].values)
plt.title('Throughput improvement. Comparing between 1 and 4 concurrent requests on single assigned CPU')
plt.ylabel('Percentage improvement requests processed')
plt.show()
```

```python
df=df.sort_values(['avg_conc_impr'], ascending=[False])
figure(num=None, figsize=(16, 6))

y_pos = np.arange(len(frameworks))
plt.bar(y_pos, df['avg_conc_impr'].values, align='center', alpha=0.5)
plt.xticks(y_pos, df['framework'].values)
plt.title('Response time improvement. Comparing between 1 and 8 assigned CPUs (no concurrency)')
plt.ylabel('Percentage improvement avg response time')
plt.show()
```

```python
df=df.sort_values(['avg_impr'], ascending=[False])
figure(num=None, figsize=(16, 6))

y_pos = np.arange(len(frameworks))
plt.bar(y_pos, df['avg_impr'].values, align='center', alpha=0.5)
plt.xticks(y_pos, df['framework'].values)
plt.title('Response time improvement. Comparing between 1 CPU no concurrency with 4 assigned CPUs 4 concurrent requests')
plt.ylabel('Percentage improvement avg response time')
plt.show()
```

```python
df=df.sort_values(['tru_impr'], ascending=[False])
figure(num=None, figsize=(16, 6))

y_pos = np.arange(len(frameworks))
plt.bar(y_pos, df['tru_impr'].values, align='center', alpha=0.5)
plt.xticks(y_pos, df['framework'].values)
plt.title('Throughput improvement. Comparing between 1 CPU no concurrency with 4 assigned CPUs 4 concurrent requests')
plt.ylabel('Percentage improvement requests processed')
plt.show()
```
